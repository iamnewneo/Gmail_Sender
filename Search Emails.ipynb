```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
comp_names = ["Pantaloons"]
```

```python
my_headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'}
#comp_names = ["freecharge"]
email_dict = {}
url = 'http://www.google.co.in/search'

regex = re.compile(("([a-z0-9!#$%&'*+\/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+\/=?^_`"
                    "{|}~-]+)*(@|\sat\s)(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?(\.|"
                    "\sdot\s))+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?)"))
#my_headers = {'User-agent' : 'Mozilla/11.0'}
```

```python
count = 1
total = len(comp_names)
for com_name in comp_names: 
    print(str(count) + " / " + str(total) + " company started.")
    #########################################################
    ################# Google Search Here  ###################
    ################# For Main Page       ###################
    ######################################################### 
    try:
        payload = {'q' : com_name, 'start' : '0'}
        r = requests.get(url, params = payload, headers = my_headers)
        soup = BeautifulSoup(r.text, 'html.parser')
        main_page = soup.find('h3',{"class" : "r"}).find('a').get('href')
        #link = re.search('url\?q=(.+?)\&sa', href).group(1)
        #print(main_page)
        
        #########################################################
        ################# Main Page  Code     ###################
        ######################################################### 
        
        page_req = requests.get(main_page, headers = my_headers)
        page_soup = BeautifulSoup(page_req.text, 'html.parser')
        print(page_soup)
        all_links = page_soup.findAll("a")
        try:
            #print("Main Page Emails")
            emails_on_page = re.findall(regex, page_req.text)
            #email_dict[com_name] = emails_on_page
            # for email in emails_on_page:
            #     print(email)
        except:
            print("No emails on Main Page")
            pass
        
        
        #########################################################
        ################# Contact Page  Code     ###################
        #########################################################
        print(all_links)
        for l in all_links:
            try:
                if "contact" in l.get("href").lower():
                    #print("Contact")
                    if str(main_page) in str(l.get("href")):
                        contact_link = str(l.get("href"))
                    else:
                        if str(l.get("href"))[0] == "/" and str(main_page)[-1] == "/":
                            contact_link = str(main_page) + str(l.get("href"))[1:]
                        else:
                            contact_link = str(main_page) + str(l.get("href"))
                    print(main_page)
                    print(contact_link)
                    contact_req = requests.get(contact_link, headers = my_headers)
                    contact_soup = BeautifulSoup(contact_req.text, 'html.parser')
                    #print(contact_soup.get_text())
                    #emails_on_contacts = re.findall(regex, contact_soup.get_text())
                    emails_on_contacts = list((email[0] for email in re.findall(regex, contact_soup.get_text()) if not email[0].startswith('//')))
                    #email_dict[com_name] = email_dict[com_name] + emails_on_contacts
                    email_dict[com_name] = emails_on_contacts
                    #print(emails_on_contacts)
                    # for em in emails_on_contacts:
                    #     print(em)
                    break
            except:
                #print("Error on Contact")
                pass
        #print(link)
        count+=1
    except:
        print("Error")
        count+=1
        continue
```
